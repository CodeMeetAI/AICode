```python
import json
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import csv
```

```python
results_file_root = "/home/eidf018/eidf018/s2484588-epcc/MLP/LLMMemoryEval/datasets/data/multiwoz_new"
sample_sizes = []
dialogue_lengths = []
dialogue_tokens = []

for dataset_file in sorted(os.listdir(results_file_root)):
        full_file_path = os.path.join(results_file_root, dataset_file)
        if full_file_path.endswith("inference.json"):
            with open(full_file_path, 'r', encoding='utf-8') as file:
                data = json.loads(file.read())
                sample_sizes.append(len(data))
                dialogue_length = []
                dialogue_token = [] 
                for dialogue in data:
                    dialogue_length.append((len(dialogue)-1)/2)
                    number_of_tokens = 0
                    for turn in dialogue:
                        content = turn['content']
                        tokens = re.sub(r'\W+', ' ', content).split()
                        number_of_token = len(tokens)
                        number_of_tokens+=number_of_token
                    dialogue_token.append(number_of_tokens)
                dialogue_tokens.append(dialogue_token)
                dialogue_lengths.append(dialogue_length)
                    
average_dialogue_lengths = [sum(dialogue)/ len(dialogue) for dialogue in dialogue_lengths]      
average_dialogue_tokens = [sum(dialogue)/ len(dialogue) for dialogue in dialogue_tokens]
```

```python
print(sample_sizes)
print(dialogue_lengths)
print(average_dialogue_lengths)
print(dialogue_tokens)
print(average_dialogue_tokens)
```

```python
label_file_root = "/home/eidf018/eidf018/s2484588-epcc/MLP/LLMMemoryEval/datasets/data/multiwoz_new/labels"
gt_counts = []
service_counts = []


for label_file in sorted(os.listdir(label_file_root)):
    full_file_path = os.path.join(label_file_root, label_file)
    if full_file_path.endswith("inference.json"):
        gt_count = {"A": 0, "B":0, "C":0}
        service_count ={'hotel': 0, 'train':0, 'attraction': 0, 'restaurant':0}
        with open(full_file_path, 'r', encoding='utf-8') as file:
            answers = json.loads(file.read())
            for answer in answers:
                gt = answer['gt'][0]
                option = answer['gt'].split(".")[1]
                if option in service_count:
                    service_count[option]+=1
                if gt in gt_count:
                    gt_count[gt]+=1
    gt_counts.append(gt_count)
    service_counts.append(service_count)            

average_gt_counts = {key: round(sum([gt[key] for gt in gt_counts])/len(gt_counts),3) for key in gt_counts[0].keys()}
average_service_counts = {key: round(sum([service[key] for service in service_counts])/len(service_counts),3) for key in service_counts[0].keys()}
```

```python
print(gt_counts)
print(service_counts)
print(average_gt_counts)
print(average_service_counts)
```

```python
with open('results_summary.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Sample Size', 'Average Dialogue Length', 'Average Dialogue Tokens','average_gt_counts', 'average_service_counts'])
    for i in range(len(sample_sizes)):
        writer.writerow([sample_sizes[i], average_dialogue_lengths[i], average_dialogue_tokens[i]], gt_counts[i], service_counts[i])
```
